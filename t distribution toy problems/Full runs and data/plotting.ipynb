```python
import numpy as np
import torch
from torch import nn
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tqdm import tqdm
from scipy.stats import norm
import pickle
```

## Gaussian Noise

```python
with open('Gaussian noise/mses_dict_gaussian_noise.pkl', 'rb') as f:
    mses_dict_g = pickle.load(f)

with open('Gaussian noise/mcp_results_gaussian.pkl', 'rb') as f:
    mcp_results_gaussian = pickle.load(f)

with open('Gaussian noise/mcp_results_t.pkl', 'rb') as f:
    mcp_results_t = pickle.load(f)

with open('Gaussian noise/mcp_results_lap.pkl', 'rb') as f:
    mcp_results_lap = pickle.load(f)

with open('Gaussian noise/mcp_results_mix.pkl', 'rb') as f:
    mcp_results_mix = pickle.load(f)
```

```python
# Correcting floating-point errors

mcp_results_gaussian = {float(f"{k:.2f}"): mcp_results_gaussian[k] for k in mcp_results_gaussian}
mcp_results_t = {float(f"{k:.2f}"): mcp_results_t[k] for k in mcp_results_t}
mcp_results_lap = {float(f"{k:.2f}"): mcp_results_lap[k] for k in mcp_results_lap}
mcp_results_mix = {float(f"{k:.2f}"): mcp_results_mix[k] for k in mcp_results_mix}
```

```python
alpha_values = np.arange(0.01, 1.0, 0.01)
x = np.linspace(0, 1, 100) # for dashed line

mcp_g = []
mcp_t = []
mcp_lap = []
mcp_mix = []

for alpha in alpha_values:
    alpha = np.round(alpha, 2)
    mcp_g.append(np.mean(mcp_results_gaussian[alpha]))
    mcp_t.append(np.mean(mcp_results_t[alpha]))
    mcp_lap.append(np.mean(mcp_results_lap[alpha]))
    mcp_mix.append(np.mean(mcp_results_mix[alpha]))

# alpha_values = np.arange(0, 1, 0.01)

# Plot the results
plt.figure(figsize=(7, 7))
plt.plot(1-alpha_values, mcp_g, label='Gaussian', marker='', linestyle='-')
plt.plot(1-alpha_values, mcp_t, label=r"Student's $t$-distribution", marker='', linestyle='-')
plt.plot(1-alpha_values, mcp_lap, label='Laplace', marker='', linestyle='-')
plt.plot(1-alpha_values, mcp_mix, label='Gaussian Mixture', marker='', linestyle='-')
plt.plot(x, x, linestyle="--", color="black", linewidth=1.5)

# Add labels, title, and legend
plt.xlabel(r"$1 - \alpha$")
plt.ylabel('Mean Coverage Probability')
plt.title("Gaussian Noise")
plt.legend()
plt.grid(True)
plt.show()
```
