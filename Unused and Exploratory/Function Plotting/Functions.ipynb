```python
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits as mpl
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import Axes3D
import random
```

```python
def f(x,y): return x**3+y**2
x = np.arange(-3,3,0.01)
y = np.arange(-3,3,0.01)
def unnormalised_ground_truth(x1,x2):
    """return the 'ground truth' output for a specific input location x"""
    y = (10 * np.sin(2 * x1 + x2) +
         5 * np.cos(x1 * x2) +
         np.exp(-0.1 * (x1**2 + x2**2)) +
         3 * x1**2 * np.log(1 + x2**2))
    return y
X, Y = np.meshgrid(x, y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
zs = np.array(unnormalised_ground_truth(np.ravel(X), np.ravel(Y)))
Z = zs.reshape(X.shape)
ax.plot_surface(X,Y,Z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
fig.set_figwidth(8)
fig.set_figheight(8)
plt.show()
```

```python
def fun(x, y):
    return x**2 + y

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = y = np.arange(-3.0, 3.0, 0.05)
X, Y = np.meshgrid(x, y)
zs = np.array(fun(np.ravel(X), np.ravel(Y)))
Z = zs.reshape(X.shape)

ax.plot_surface(X, Y, Z)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
